### Bar Charts and Pie Charts

```python
occuaptions = df['occupation'].unique().tolist()
occuaption_counts = df['occupation'].value_counts(sort=False)

plt.bar(occuaptions, occuaption_counts)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(occuaption_counts, labels=occuaptions, autopct='%1.1f%%', pctdistance=0.80)
plt.legend(bbox_to_anchor=(1.85,0.5), loc=5)
plt.show()
```

#### From the above graph we can clearly understand that pie charts are not always that useful.

```python
gender_count = df['gender'].value_counts()
gender_label = ['Male', 'Female']

plt.pie(gender_count, labels=gender_label, autopct='%1.1f%%')
plt.legend(title="Gender", bbox_to_anchor=(1.35, 0.5), loc='right')
plt.show()
```
